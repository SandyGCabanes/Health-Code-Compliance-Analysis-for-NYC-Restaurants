# AUTOMATED EXTRACTION AND LOADING OF RESTAURANT HEALTH INSPECTIONS NEW YORK CITY
- No need for Luigi or Airflow or even cron or Windows Task Scheduler. 
- Data set is manageable.  Manual trigger is not difficult.

```python
#  Import only the recent dates after the latest full month
import requests
import pandas as pd

base_url = "https://data.cityofnewyork.us/resource/43nn-pn8j.json"
chunk_size = 50000
offset = 0

where_clause = (
    "inspection_date <= '2026-01-20T23:59:59.000' "
    "AND inspection_date > '2025-08-31T23:59:59.000'"
)

all_chunks = []

while True:
    params = {
        "$limit": chunk_size,
        "$offset": offset,
        "$where": where_clause
    }
    r = requests.get(base_url, params=params)
    data = r.json()
    if not data:
        break
    all_chunks.append(pd.DataFrame(data))
    offset += chunk_size

df = pd.concat(all_chunks, ignore_index=True)

# Load to input_file directory
df.to_csv("input_file/DOHMH_NYC_Resto_Health_20250831_20260120.csv", index=False)

```
